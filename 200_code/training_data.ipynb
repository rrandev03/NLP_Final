```python
import pandas as pd
import numpy as np

pd.set_option("mode.copy_on_write", True)
```

```python
jan_2021 = pd.read_excel(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/101_raw/ev_jan_june_2021.xlsx"
)
june_2021 = pd.read_excel(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/101_raw/ev_june_dec_2021.xlsx"
)
```

```python
jan_2020 = pd.read_csv(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/102_processed/jan_2020_processed.csv"
)
june_2020 = pd.read_csv(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/102_processed/june_2020_processed.csv"
)
jan_2023 = pd.read_csv(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/102_processed/jan_2023_processed.csv"
)
june_2023 = pd.read_csv(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/102_processed/june_2023_processed.csv"
)
```

```python
concat = pd.concat([jan_2020, june_2020, jan_2023, june_2023])
intermed_train = concat.loc[~concat["Manual_Training_Labels"].isna()]
assert len(intermed_train) == 100
```

```python
intermed_train = intermed_train[["Date", "Tweet_Content", "Manual_Training_Labels"]]
```

```python
intermed_train = intermed_train.reset_index(drop=True)
```

```python
intermed_test = concat.loc[~concat["Manual_Testing_Labels"].isna()]
assert len(intermed_test) == 52
```

```python
intermed_test = intermed_test[["Date", "Tweet_Content", "Manual_Testing_Labels"]]
intermed_test = intermed_test.reset_index(drop=True)
```

```python
jan_2021 = jan_2021.rename(columns={"LABELS": "Labels"})
```

```python
jan_2021 = jan_2021.loc[~jan_2021["Labels"].isna()]
june_2021 = june_2021.loc[~june_2021["Labels"].isna()]
jan_2021["Date"] = pd.to_datetime(jan_2021["UTC_Time"]).dt.date
june_2021["Date"] = pd.to_datetime(june_2021["UTC_Time"]).dt.date
intermed_2021 = pd.concat([jan_2021, june_2021]).reset_index(drop=True)
intermed_2021 = intermed_2021[["Date", "Tweet_Content", "Labels"]]
assert len(intermed_2021) == 76
```

```python
test_2021 = intermed_2021.drop(train_indices)
test_2021 = test_2021.reset_index(drop=True)
assert len(test_2021) == 26
```

```python
train_2021.loc[train_2021["Labels"] == "POS", "Labels"] = "POSITIVE"
```
