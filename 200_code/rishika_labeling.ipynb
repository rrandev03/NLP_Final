```python
import pandas as pd
import numpy as np

pd.set_option("mode.copy_on_write", True)
```

```python
jan_2020 = pd.read_csv(
    "https://github.com/rrandev03/NLP_Final/raw/refs/heads/main/100_data/101_raw/ev_jan_june_2020.csv"
)
jan_2020.sample()
```

# Training Labels for Jan - June 2020

```python
train = jan_2020.sample(25, random_state=30).index
train
```

```python
pd.set_option("display.max_colwidth", None)
jan_2020.iloc[train, :][["UTC_Time", "Tweet_Content"]]
```

```python
train_manual_labels = [
    "Positive",
    "Positive",
    "Neutral",
    "Neutral",
    "Neutral",
    "Neutral",
    "Neutral",
    "Neutral",
    "Positive",
    "Neutral",
    "Neutral",
    "Neutral",
    "Positive",
    "Neutral",
    "Neutral",
    "Negative",
    "Neutral",
    "Positive",
    "Neutral",
    "Positive",
    "Neutral",
    "Positive",
    "Positive",
    "Neutral",
    "Neutral",
]
assert len(train_manual_labels) == 25
```

```python
jan_2020.loc[train, "Manual_Training_Labels"] = train_manual_labels
```

```python
jan_2020["Is_Training"] = jan_2020["Manual_Training_Labels"].isna().astype(int)
```

```python
jan_2020.loc[train][["Tweet_Content", "Is_Training", "Manual_Training_Labels"]]
```

# Testing Labels for Jan - June 2020

```python
test = jan_2020.sample(13, random_state=40).index
test
```

```python
assert len(set(train).intersection(set(test))) == 0  # Make sure no overlap
```

```python
pd.set_option("display.max_colwidth", None)
jan_2020.iloc[test, :][["UTC_Time", "Tweet_Content"]]
```

```python
test_manual_labels = [
    "Neutral",
    "Neutral",
    "Neutral",
    "Positive",
    "Positive",
    "Neutral",
    "Positive",
    "Neutral",
    "Neutral",
    "Neutral",
    "Positive",
    "Neutral",
    "Neutral",
]
assert len(test_manual_labels) == 13
```

```python
jan_2020.loc[test][["Tweet_Content", "Is_Testing", "Manual_Testing_Labels"]]
```
