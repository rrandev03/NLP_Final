```python
import os
import pandas as pd

# Specify the folder containing the files
folder_path = "../100_data/101_raw"

# Get a list of all CSV and Excel files except 'synthetic_tweets.csv'
csv_files = [
    file
    for file in os.listdir(folder_path)
    if file.endswith(".csv")
    and file != "synthetic_tweets.csv"
    and file != "Jan-June 2022 Tweets_labeled.xlsx - Sheet1.csv"
    and file != "July-Dec 2022_labeled.xlsx - Sheet1.csv"
]
excel_files = [
    file for file in os.listdir(folder_path) if file.endswith((".xlsx", ".xls"))
]

# Combine only the required columns
combined_df = pd.DataFrame()

# Process CSV files
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    # Read only the "Tweet_Content" and "UTC_Time" columns
    df = pd.read_csv(file_path, usecols=["Tweet_Content", "UTC_Time"])
    combined_df = pd.concat(
        [combined_df, df], ignore_index=True
    )  # Concatenate dataframes

# Process Excel files
for excel_file in excel_files:
    file_path = os.path.join(folder_path, excel_file)
    # Read only the "Tweet_Content" and "UTC_Time" columns
    df = pd.read_excel(file_path, usecols=["Tweet_Content", "UTC_Time"])
    combined_df = pd.concat(
        [combined_df, df], ignore_index=True
    )  # Concatenate dataframes

# Save the combined dataframe to a new CSV file in ../100_data/101_raw
output_path = os.path.join(folder_path, "combined_file.csv")
combined_df.to_csv(output_path, index=False)

print(
    f"Combined {len(csv_files) + len(excel_files)} files (excluding 'synthetic_tweets.csv') and fan's files into '{output_path}'"
)
```

```python
import os
import pandas as pd

# Specify the folder containing the files
folder_path = "../100_data/101_raw"

# Get a list of all CSV and Excel files
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]
excel_files = [
    file for file in os.listdir(folder_path) if file.endswith((".xlsx", ".xls"))
]


# Function to check column existence
def check_columns(file_path, required_columns, file_type):
    try:
        if file_type == "csv":
            df = pd.read_csv(file_path, nrows=0)  # Read only the header
        elif file_type == "excel":
            df = pd.read_excel(file_path, nrows=0)  # Read only the header
        else:
            return False

        # Check if all required columns exist
        missing_columns = [col for col in required_columns if col not in df.columns]
        if missing_columns:
            print(f"File '{file_path}' is missing columns: {missing_columns}")
        else:
            print(f"File '{file_path}' has all required columns: {required_columns}")

    except Exception as e:
        print(f"Error reading file '{file_path}': {e}")


# Columns to check
required_columns = ["UTC_Time", "Tweet_Content"]

# Check CSV files
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    check_columns(file_path, required_columns, file_type="csv")

# Check Excel files
for excel_file in excel_files:
    file_path = os.path.join(folder_path, excel_file)
    check_columns(file_path, required_columns, file_type="excel")
```
